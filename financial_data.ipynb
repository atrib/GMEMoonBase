```python
import yfinance as yf

#define the ticker symbol
tickerSymbol = 'GME'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
tickerDf = tickerData.history(period='1d', start='2019-1-1', end='2021-3-11')

#see your data
tickerDf
```

```python
import matplotlib.pyplot as plt

tickerDf['median_daily'] = (tickerDf.High+tickerDf.Low)/2
print(type(tickerDf.index))

plt.figure(figsize=(20,10))
plt.title('Game Stop')
plt.plot(tickerDf.index, tickerDf['median_daily'])
```
